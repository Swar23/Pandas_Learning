```python
import pandas as pd
import numpy as np

sales = [40,10,30,20,50]
ser = pd.Series(sales, index=['day3','day1','day4','day2','day5'], name='numbers')
print(ser)
print('*' * 30)

# Sorting in pandas series can be done by both values and index

# Sort by values
print(f'sort values --- {ser.sort_values()}')
print('-' * 30)
print(f'sort values descending --- {ser.sort_values(ascending = False)}')
print('*' * 30)

#sort by index
print(f'sort index --- {ser.sort_index()}')
print('-' * 30)
print(f'sort index descending --- {ser.sort_index(ascending = False)}')
print('*' * 30)

# sort inplace
ser.sort_values(inplace=True)
print(f'inplace sort values---{ser}')
print('*' * 30)

# sorting with NaN
ser_nan = pd.Series([40,10,np.nan,30,20])
print(f'sort with nan---{ser_nan.sort_values()}')
print('-' * 30)
print(f'sort with nan---{ser_nan.sort_values(na_position = 'first')}')
```

```python
# Numerical series aggregation

sales = pd.Series([10,20,30,40,50], name='numbers')

print(sales)
print('*' * 30)

# 1) Sum
print(f'sum is -- {sales.sum()}')
print('*' * 30)

# 2) prod
print(f'prod is -- {sales.prod()}')
print('*' * 30)

# 3) mean and median
print(f'mean is -- {sales.mean()}')
print('-' * 30)
print(f'median is -- {sales.median()}')
print('*' * 30)

# 4) min ,max ,argmin,argmax
print(f'min is -- {sales.min()}')
print('-' * 30)
print(f'max is -- {sales.max()}')
print('-' * 30)
print(f'argmin is -- {sales.argmin()}')
print('-' * 30)
print(f'argmax is -- {sales.argmax()}')
print('*' * 30)

# 5)std and var
print(f'std---{sales.std()}')
print('-' * 30)
print(f'var---{sales.var()}')
print('*' * 30)

# 6) count (non-NaN values)
print(f'count---{sales.count()}')
print('*' * 30)

# 7) quantile
print(f'quantile---{sales.quantile(0.5)}')
print('-' * 30)
# interpolation  lower,higher,nearest,linear(default),midpoint
print(f'quantile---{sales.quantile(0.5,interpolation = 'lower')}')
print('-' * 30)

```

```python
# categorical series aggregation

category_s = pd.Series(
    ['Low','Medium','High','Low','High','Low','Medium'],
    dtype='category'
)

print(category_s)
print('*' * 30)

# 1) count (non-null values)
print(f'Count -- {category_s.count()}')
print('*' * 30)

# 2) value_counts (most important)
print(f'Values_Counts -- {category_s.value_counts()}')
print('*' * 30)

# 3) mode (most frequent value)
print(f'mode---{category_s.mode()}')
print('*' * 30)

# 4) unique categories present
print(f'unique---{category_s.unique()}')
print('*' * 30)

# 5) number of unique categories
print(f'nunique---{category_s.nunique()}')

print(category_s.describe())
```

```python
series_d = pd.Series(np.arange(1,21), name='numbers')

print(series_d)
print('*' * 30)

# 1) head() → first 5 values (default)
print(f'head default---{series_d.head()}')
print('*' * 30)

# head(n)
print(f'head 3---{series_d.head(3)}')
print('*' * 30)

# 2) tail() → last 5 values (default)
print(f'tail default---{series_d.tail()}')
print('*' * 30)

# tail(n)
print(f'tail 3---{series_d.tail(3)}')
print('*' * 30)

# 3) sample() → random rows
print(f'sample 4---{series_d.sample(4)}')
print('-' * 30)
print(f'sample random state---{series_d.sample(4,random_state = 42)}')
print('-' * 30)

# 4) info()
print(series_d.info())
```

```python
# missing value


s = pd.Series([10, 20, np.nan, 40, None, 60], name='numbers')

print(s)
print('*' * 30)

# 1)isna to check for missing value
print(f'isna --- {s.isna()}')
print('*' * 30)

# 2) dropna to drop the missing value
print(f'dropna ---- {s.dropna()}')
print('*' * 30)

s = pd.Series([10, 20, np.nan, 40, None, 60], name='numbers')

# 3) fillna to fill the nan value
print(f'fillna ---- {s.fillna(10)}')
print('*' * 30)
```

```python
# apply lets you apply custom function to panda series
s = pd.Series([10, 20, 30, 40])

result = s.apply(lambda x: x * 2)
print(result)
print('*' * 30)

# where(condition,value if false , inplace = True/False)
result = s.where(s > 20, -1)
print(result)
print('*' * 30)

#mask(condition,value if true)
result = s.mask(s > 20, -1)
print(result)
print('*' * 30)
```

```python
# map If value not present in dictionary → it becomes NaN,Works only on Series
s = pd.Series(['A','B','C','A'])

mapping = {
    'A': 'Apple',
    'B': 'Banana',
    'C': 'Cherry'
}

print(s.map(mapping))

s = pd.Series([10,20,30])
# using function
print(s.map(lambda x: x * 2))
```
