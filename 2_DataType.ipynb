```python
import pandas as pd
import numpy as np

# Series numpy datatype
# int64
series = pd.Series([1,2,3,4,5,6,7,8],name = 'rollNo')
print(series.dtype)

# int64 numpy datatype, it is not able to handle missing values hence it converts int64 to float,missing value np.nan
series_miss = pd.Series([1,2,3,4,np.nan])
print(series_miss.dtype)

#float64,same as int64 in case of missing value
series_f = pd.Series([1,2,3,4,5],name= 'float',dtype = 'float64')
print(series_f.dtype)

# bool , if we add missing value the datatype becomes object
series_bool = pd.Series([True,False,True])
print(series_bool.dtype)

# contains data of multiple type,if we have only strings in that case also it will object dtype, missing value is np.nan
series_object = pd.Series([1,'test',True,20.45])
print(series_object.dtype)

# no seperate string type ,strings are treated as object
series_str = pd.Series(['and','sad','got'])
print(series_str.dtype)
```

```python
# Pandas datatype

# Int64(nullable int), it is better at handling missing value then int64 not converted to float, missing value is <NA>
pandas_int = pd.Series([1,2,3,4,np.nan],dtype = 'Int64')
print(pandas_int)

# Float64(nullable float), it is better at handling missing value then float64, missing value is <NA>
pandas_float = pd.Series([1.0,2.0,3.0,4.0,np.nan],dtype = 'Float64')
print(pandas_float)

# boolean,missing values handled better than bool and not converted to object in case it contains missing value
pandas_bool = pd.Series([True,False,True,np.nan],dtype = 'boolean')
print(pandas_bool)

#String ,a seperate datatype to deal with strings better for working with string using str(we learn this later)
pandas_string = pd.Series(['and','sad','got'],dtype='string')
print(pandas_string.dtype)


# Category,used when series has repeated values from limited set
pandas_category = pd.Series(['Adult','Teen','Child','Teen','Adult'],dtype = 'category')
print(pandas_category)

# internally it is stored as codes and categories rather than storing the values as it is ,this helps save memory
print(f'codes ---- {pandas_category.cat.codes}')
print(f'categories ---- {pandas_category.cat.categories}')

# We can add categories
pandas_category = pandas_category.cat.add_categories(['Senior'])
print(f'categories new added ---- {pandas_category.cat.categories}')

# similar way we can remove categories using remove_categories() and remove_unused_categories()
pandas_category = pandas_category.cat.remove_categories(['Teen'])
print(f'categories new removed ---- {pandas_category.cat.categories}')
# removed categories is replaced by NaN in final result
print(pandas_category)

# ordering category ,with this words become ranked level ,and default python comparision with words just as label which is of no use in real world would be replaced
pd_order = pd.Series(['High','Low','Medium'],dtype = pd.CategoricalDtype(
    categories=['High','Low','Medium'],
    ordered= True
))
print(pd_order)
```

```python
# DateTime

# datetime64 ,a single moment in time eg - (January 4 2016,2:00:00pm)
pd_datetime = pd.Series(pd.to_datetime([
    '2026-04-04',
    '2026-04-06',
    '2026-04-14',
    '2026-04-24',
]))
print(pd_datetime)
print('*' * 30)

#timedelta ,duration between two dates
pd_timedelta = pd_datetime - pd_datetime[0]
print(pd_timedelta)

# add 7 days
new_dates = pd_datetime + pd.Timedelta(days=7)
print(new_dates)
print('*' * 30)

#Period ,a span of time i.e week ,month
monthly_period = pd_datetime.dt.to_period("M")
print(monthly_period)


```
