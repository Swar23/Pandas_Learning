```python
import pandas as pd

df1 = pd.DataFrame({
    "id": [1,2,3],
    "Name": ["A","B","C"]
})

df2 = pd.DataFrame({
    "id": [2,3,4],
    "Salary": [20000,30000,40000]
})

# inner join
df3 = pd.merge(df1,df2 ,on = 'id', how = 'inner')
df3
```

```python
# left join ,similar for right and outer
df4 = pd.merge(df1,df2 , on ='id',how ='left')
df4
```

```python
# join based on multiple col
df1 = pd.DataFrame({
    "id": [1,2,3],
    "Name": ["A","B","C"],
    "City": ["Mumbai","Pune","Delhi"]
})

df2 = pd.DataFrame({
    "id": [1,2,3],
    "Name": ["A","B","X"],
    "Salary": [20000,30000,40000]
})

pd.merge(df1, df2, on=["id","Name"], how="inner")

```

```python
# concat
df3 = pd.concat([df1, df2], axis=0)  # row-wise
df3
```

```python
df_c = pd.concat([df1, df2], axis=1)  # column-wise
df_c
```

```python
# join with tables having same column name
ds = df1.set_index("id").join(
   df2.set_index("id"),
   how="inner",
   lsuffix="_left",
   rsuffix="_right"
)

ds
```
