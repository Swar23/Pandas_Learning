```python
import pandas as pd
import numpy as np

sales = [10,20,30,40,50,60]
ser = pd.Series(sales,name = 'numbers')

# Filtering series

# 1) ==(python) or eq(pandas)
print(f'equal---{ser[ser == 30]}')
print('-' * 30)
print(f'equal pandas---{ser[ser.eq(30)]}')

print('*' * 30)
# 2) != or ne()
print(f'not equal---{ser[ser != 30]}')
print('-' * 30)
print(f'not equal pandas---{ser[ser.ne(30)]}')
print('*' * 30)

# 3) < or lt
print(f'less than---{ser[ser < 30]}')
print('-' * 30)
print(f'less than pandas---{ser[ser.lt(30)]}')
print('*' * 30)

# 4) <= or le
print(f'less than equal---{ser[ser <= 30]}')
print('-' * 30)
print(f'less than equal pandas---{ser[ser.le(30)]}')
print('*' * 30)

# 5) Similarly we can apply for  >/gt and >=/ge

# 6) membership test isin and ~isin
print(f'isin pandas---{ser[ser.isin([10,20,30])]}')
print('-' * 30)
print(f'~isin pandas---{ser[~ser.isin([30])]}')
print('*' * 30)

# 7) combine condition &(and) , | (or)
print(f'Combined condition and--- {ser[(ser > 10) & (ser < 50)]}')
print('-' * 30)
print(f'Combined condition or--- {ser[(ser > 10) | (ser < 50)]}')
```

```python
import pandas as pd
import numpy as np

sales = [10,20,30,40,50,60]
ser = pd.Series(sales,name = 'numbers')

# Arithmetic operations on series

# 1) +(python) or add(pandas)
print(f'add---{ser + 5}')
print('-' * 30)
print(f'add pandas---{ser.add(5)}')
print('*' * 30)

# 2) -(python) or sub()
print(f'subtract---{ser - 5}')
print('-' * 30)
print(f'subtract pandas---{ser.sub(5)}')
print('*' * 30)

# 3) *(python) or mul()
print(f'multiply---{ser * 2}')
print('-' * 30)
print(f'multiply pandas---{ser.mul(2)}')
print('*' * 30)

# 4) /(python) or div()
print(f'divide---{ser / 2}')
print('-' * 30)
print(f'divide pandas---{ser.div(2)}')
print('*' * 30)

# 5) //(python) or floordiv()
print(f'floor divide---{ser // 3}')
print('-' * 30)
print(f'floor divide pandas---{ser.floordiv(3)}')
print('*' * 30)

# 6) %(python) or mod()
print(f'modulus---{ser % 4}')
print('-' * 30)
print(f'modulus pandas---{ser.mod(4)}')
print('*' * 30)

# 7) **(python) or pow()
print(f'power---{ser ** 2}')
print('-' * 30)
print(f'power pandas---{ser.pow(2)}')
print('*' * 30)

# when we use pandas method we can use fillvalue when a value is missing
ser1 = pd.Series([10,20,30], index=['a','b','c'])
ser2 = pd.Series([1,2,3], index=['a','b','d'])

print(ser1.add(ser2, fill_value=0))
print('-' * 30)
ser_demo = pd.Series([1,2,3,np.nan])
ser_add  = ser_demo.add(2, fill_value = 0)
print(ser_add)
```

```python
# Text series Operator

text_s = pd.Series([
    "Coffee",
    "tea",
    "Green Tea",
    "black coffee",
    "TEA"
], name='drinks')

print(text_s)
print('*' * 30)

# we need to use .str to work with the below methods

# 1) Case conversion upper ,lower ,title
print(f'upper text --- {text_s.str.upper()}')
print('-' * 30)
print(f'lower text --- {text_s.str.lower()}')
print('-' * 30)
print(f'lower title --- {text_s.str.title()}')
print('*' * 30)

# 2) contains , case = False to ignore case sensitivity
print(f'contains ---- {text_s.str.contains('tea',case = False)}')
print('*' * 30)

# 3) startswith and endswith
print(f'startswith ---- {text_s.str.startswith('tea')}')
print('-' * 30)
print(f'endswith ---- {text_s.str.endswith('ee')}')
print('*' * 30)

# 4)replace
print(f'replace ---- {text_s.str.replace('coffee','latte',case = False)}')
print('*' * 30)

# 5) strip
space_s = pd.Series(["  tea  ", " coffee ", "milk  "])
print(f'strip -- {space_s}')
print(f'strip -- {space_s.str.strip()}')
print('-' * 30)
print(f'lstrip---{space_s.str.lstrip()}')
print('-' * 30)
print(f'rstrip---{space_s.str.rstrip()}')
print('*' * 30)

# 6) count characters
print(f'count char --- {space_s.str.count('e')}')
print('*' * 30)

# 7) extract using regex
print(f'extract Tea---{text_s.str.extract(r"(Tea)")}')
print('*' * 30)

# 8) split
print(f'split---{text_s.str.split(" ",expand = True)}')

# 9) length of string
print(f'length---{text_s.str.len()}')
```
