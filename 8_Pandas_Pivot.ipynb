```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['IT', 'HR', 'IT', 'HR', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Salary': [50000, 40000, 60000, 45000, 70000]
})

print(df)
```

```python
#Groupby single column
print(df.groupby('Department')['Salary'].mean())
print('-' * 30)

# group by multi column,this creates index for each column
print(df.groupby(['Department', 'Employee'])['Salary'].sum())
print('-' * 30)

# group by multi column,this does not creates index for each column
print(df.groupby(['Department', 'Employee'],as_index = False)['Salary'].sum())
print('-' * 30)

# multiple aggregations
print(df.groupby('Department')['Salary'].agg(['sum','mean']))
print('-'*30)

print(df.groupby('Department')['Salary'].mean().reset_index())
print('-'*30)

# Group entire Df
print(df.groupby('Department').sum(numeric_only=True))
print('-'*30)

# custom function
print(df.groupby('Department')['Salary'].apply(lambda x: x.max() - x.min()))
print('-'*30)

```

```python
df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR'],
    'Year': [2023, 2024, 2023, 2024],
    'Salary': [50000, 60000, 40000, 45000]
})

# multilevel index Level 0 → Department Level 1 → Year
grouped = df.groupby(['Department', 'Year'])['Salary'].sum()
print(grouped)

# 1) Swaplevel  Level 1 → Department Level 0 → Year
print('-' * 30)
print(grouped.swaplevel())
# grouped.swaplevel(0, 1)

# drop level
print('-' * 30)
grouped.droplevel('Year')
```

```python
df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR'],
    'Salary': [50000, 60000, 40000, 45000],
    'Bonus': [5000, 7000, 3000, 3500]
})

# agg demo
print(df.groupby('Department').agg({
    'Salary': 'mean',
    'Bonus': 'sum'
}))
print('-' * 30)

# multi func per col
print(df.groupby('Department').agg({
    'Salary': ['mean', 'max'],
    'Bonus': ['sum', 'min']
}))
print('-' * 30)

# named agg
print(df.groupby('Department').agg(
    Avg_Salary = ('Salary', 'mean'),
    Total_Bonus = ('Bonus', 'sum')
))
print('-' * 30)

# custom func
df.groupby('Department')['Salary'].agg(
    lambda x: x.max() - x.min()
)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Sales"],
    "Employee": ["A", "B", "C", "D", "E"],
    "Salary": [50000, 60000, 40000, 45000, 70000]
})

print(df)
print('-' * 30)
print(df.groupby("Department")["Salary"].mean())
print('-' * 30)
# Transform works with groupby and helps keep the same number of row
df["Dept_Avg_Salary"] = df.groupby("Department")["Salary"].transform("mean")

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Sales", "Sales"],
    "Gender": ["M", "F", "M", "F", "M", "F"],
    "Salary": [50000, 60000, 40000, 45000, 70000, 65000]
})

print(df)
print('-' * 30)

#Average salary per dept
print(df.pivot_table(values = 'Salary',index = 'Department',aggfunc= 'mean'))
print('-' * 30)

# Avg salary per dept based on gender
print(df.pivot_table(values = 'Salary',columns='Gender',index = 'Department',aggfunc = 'mean'))
print('-' * 30)

# fill missing value
print(df.pivot_table(
    values="Salary",
    index="Department",
    columns="Gender",
    aggfunc="mean",
    fill_value=0
))
print('-' * 30)

df.pivot_table(
    values="Salary",
    index="Department",
    columns="Gender",
    aggfunc="mean",
    margins=True
)
```

```python
# heatmap creation using gradient ,we can use seaborn or matplotlib library as well for this
import pandas as pd

df = pd.DataFrame({
    "Math": [90, 70, 80, 60],
    "Science": [85, 75, 65, 95],
    "English": [88, 72, 78, 85]
})

# higher -- darker , lower -- lighter , default is Blues
df.style.background_gradient(cmap = 'Greens')



# apply to specific col
df.style.background_gradient(subset=["Math","Science"],cmap = 'Reds')
```

```python
# melt to convert wide table to long table
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B"],
    "Math": [90, 70],
    "Science": [85, 75],
    "English": [88, 72]
})

melted = df.melt(id_vars="Name")
print(melted)
print('-'* 30)

melt1 = df.melt(id_vars = 'Name',value_vars = ['Math','Science'],var_name = 'Subject',value_name='Score')
print(melt1)
```
