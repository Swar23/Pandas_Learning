```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D','E'],
    'Age': [20, 25, 30, 35,pd.NA],
    'Salary': [30000, 20000, 40000, 50000,60000]
})

print(df)
```

```python
# Query
# 1) Simple filter by age
data = df.query('Age > 25')
print(f'simple query --- {data}')
print('-' * 30)

# multiple condition (and)
data = df.query('Age > 32 and Salary >30000')
print(data)
print('-' * 30)

# multiple condition (or)
data = df.query('Age > 32 or Salary >30000')
print(f'OR---{data}')
print('-' * 30)

# using variables in query
salary = 40000
print(f'External Variable --- {df.query('Salary > @salary')}')
print('-' * 30)

```

```python
#sort

data = df.sort_values('Age',ascending = False)
print(f'{data}')
print('-' * 30)

# Multiple columns (First sort by age if age same then sort by salary)
data = df.sort_values(['Age','Salary'])
print(f'{data}')
print('-' * 30)

# Different order for each column
data = df.sort_values(['Age','Salary'],ascending = [False,True])
print(f'{data}')
print('-' * 30)

print(f'Sor -- {df.sort_values('Age', na_position='first')}')
print('-' * 30)
# sort by index
data = df.sort_index(ascending = [False])
print(f'{data}')
print('-' * 30)

data = df.sort_index(axis=1)
print(f'{data}')
print('-' * 30)
```

```python
#Rename columns ,can have , inplace=True if we want to change df
data = df.rename(columns={'Name': 'Employee_Name', 'Age': 'Employee_Age'})
print(f'{data}')
print('-' * 30)


df.columns = ['Emp_Name', 'Emp_Age', 'Emp_Salary']
print(f'{df}')
print('-' * 30)

data = df.rename(columns=lambda x: x.replace("_", "-"))
print(f'{data}')
print('-' * 30)

data = df.rename(index={0: 'row1', 1: 'row2'})
print(f'{data}')
print('-' * 30)
```

```python
import numpy as np
import pandas as pd

ages = pd.Series([15, 22, 35, 12, 60])

conditions = [
    ages < 18,
    ages < 40,
    ages >= 40
]

choices = ['Teen', 'Adult', 'Senior']

# select similar to if-else structure
result = np.select(conditions, choices, default='Unknown')

print(result)
```

```python
# map If value not present in dictionary → it becomes NaN,Works only on Series
s = pd.Series(['A','B','C','A'])

mapping = {
    'A': 'Apple',
    'B': 'Banana',
    'C': 'Cherry'
}

print(s.map(mapping))

s = pd.Series([10,20,30])
# using function
print(s.map(lambda x: x * 2))
```

```python
# assign ,to create new columns
import pandas as pd

df = pd.DataFrame({
    'Name': ['A', 'B', 'C'],
    'Salary': [30000, 40000, 50000]
})

new_df = df.assign(Bonus=5000)
print(new_df)
print('-' * 30)

# create col using existing col
new_df = df.assign(Bonus = df['Salary'] * 0.6)
print(new_df)
print('-' * 30)

# using lambda
df2 = df.assign(Bonus = df['Salary'] * 0.5 , Tax = lambda x : x['Bonus'] / 30)
print(df2)
print('-' * 30)


# chaining
demo = df.assign(
    Bonus=lambda x: x['Salary'] * 0.1
).query("Bonus > 3000").sort_values("Bonus")
print(demo)
```
