```python
# apply lets you apply custom function to panda series
s = pd.Series([10, 20, 30, 40])

result = s.apply(lambda x: x * 2)
print(result)
print('*' * 30)

# where(condition,value if false , inplace = True/False)
result = s.where(s > 20, -1)
print(result)
print('*' * 30)

#mask(condition,value if true)
result = s.mask(s > 20, -1)
print(result)
print('*' * 30)
```

```python
# map If value not present in dictionary → it becomes NaN,Works only on Series
s = pd.Series(['A','B','C','A'])

mapping = {
    'A': 'Apple',
    'B': 'Banana',
    'C': 'Cherry'
}

print(s.map(mapping))

s = pd.Series([10,20,30])
# using function
print(s.map(lambda x: x * 2))
```

```python
# working with duplicates
import pandas as pd

df = pd.DataFrame({
    'Name': ['A', 'B', 'A', 'C', 'B'],
    'Age': [20, 25, 20, 30, 25]
})

print(df)
print('-'*30)
print(df.duplicated())
print('-'*30)
print(df.duplicated(subset='Name'))
print('-'*30)
print(df.duplicated(subset=['Name', 'Age']))
print('-'*30)
df.duplicated(keep='last') #Last occurrence = False Earlier ones = True ,first is other value
print('-'*30)
df.duplicated(keep=False) #all duplicates are true,else false
print('-'*30)
df.drop_duplicates(subset='Name')
```
